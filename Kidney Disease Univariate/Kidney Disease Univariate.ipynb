```python
def quanQual(dataset):
        nominal_cols = []
        ordinal_cols = []
        for columnName in dataset.columns:
            if(dataset[columnName].dtype == 'O'):
                nominal_cols.append(columnName)
            else:
                ordinal_cols.append(columnName)
        return nominal_cols, ordinal_cols
```

```python
nominal_cols, ordinal_cols = quanQual(dataset)
```

```python
#First Step
#Check how many missing values each column has
print(dataset.isnull().sum())
```

```python
# Impute ordinal columns with median using SimpleImputer
from sklearn.impute import SimpleImputer

ordinal_imputer = SimpleImputer(strategy='median')
dataset[ordinal_cols] = ordinal_imputer.fit_transform(dataset[ordinal_cols])
```

```python
#for nominal (categorical) columns: use mode (most frequent)
for columnName in nominal_cols:
    dataset[columnName] = dataset[columnName].fillna(dataset[columnName].mode()[0])
```

```python
def Univariate(dataset, ordinal_cols):
    descriptive = pd.DataFrame(index = ["Mean", "Median", "Mode","Q1:25%","Q2:50%","Q3:75%","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max"], columns=ordinal_cols)
    for columnName in ordinal_cols:
        descriptive.loc["Mean",columnName] = dataset[columnName].mean()
        descriptive.loc["Median",columnName] = dataset[columnName].median()
        descriptive.loc["Mode",columnName] = dataset[columnName].mode()[0]
        descriptive.loc["Q1:25%",columnName] = dataset.describe()[columnName]["25%"]
        descriptive.loc["Q2:50%",columnName] = dataset.describe()[columnName]["50%"]
        descriptive.loc["Q3:75%",columnName] = dataset.describe()[columnName]["75%"]
        descriptive.loc["Q4:100%",columnName] = dataset.describe()[columnName]["max"]
        descriptive.loc["IQR",columnName] = (descriptive.loc["Q3:75%", columnName] - descriptive.loc["Q1:25%", columnName])
        descriptive.loc["1.5rule",columnName] =1.5 * descriptive[columnName]["IQR"]
        descriptive.loc["Lesser",columnName] = descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5rule"]
        descriptive.loc["Greater",columnName] = descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5rule"]
        descriptive.loc["Min",columnName] = dataset[columnName].min()
        descriptive.loc["Max",columnName] = dataset[columnName].max()
    return descriptive  
            
```

```python
descriptive = Univariate(dataset, ordinal_cols)
```

```python
descriptive
```

```python
lesser = []
greater = []

for columnName in ordinal_cols:
    if(descriptive[columnName]["Min"] < descriptive[columnName]["Lesser"]):
        lesser.append(columnName)
    if(descriptive[columnName]["Max"] > descriptive[columnName]["Greater"]):
        greater.append(columnName)
```

```python
for columnName in lesser:
    dataset[columnName][dataset[columnName] < descriptive[columnName]["Lesser"]] = descriptive[columnName]["Lesser"]
for columnName in greater:
    dataset[columnName][dataset[columnName] > descriptive[columnName]["Greater"]] = descriptive[columnName]["Greater"]
```
